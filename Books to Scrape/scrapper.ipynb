This is a demo web scrapper which scrapes book titles from the website "Books to Scrape"<br> 

URL to the website used : <a href="https://books.toscrape.com">Books to Scrape</a>

```python
import requests
import bs4
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
#Using format method
page_num = 23
base_url.format(page_num)
```

```python
res = requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
products = soup.select('.product_pod')
```

```python
example = products[0]
```

```python
example
```

```python
#A quick way to check if a substring is present in a bigger string
'star-rating Three' in str(example)
```

```python
#Better method to check the book rating
example.select('.star-rating.Three')
```

```python
example.select('.star-rating.Two') == []
```

```python
#Grabbing the title of the book
example.select('a')[1]['title']
```

```python
recommended_titles = []

for num in range(1,51):
    res = requests.get(base_url.format(num))
    soup = bs4.BeautifulSoup(res.text,'lxml')
    books = soup.select('.product_pod')
    for book in books:
        if book.select('.star-rating.Four') != [] or book.select('.star-rating.Five') != []:
            title = book.select('a')[1]['title']
            recommended_titles.append(title)
```
